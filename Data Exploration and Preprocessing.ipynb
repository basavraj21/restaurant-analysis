```python
#1 Create a map centered at the average latitude and longitude
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)
```

```python
# Add restaurant locations to the map
for _, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name'],  
        tooltip=row['Aggregate rating']
    ).add_to(restaurant_map)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv')
```

```python
#2(a) Distribution of restaurants by city
city_distribution = df['City'].value_counts()
plt.figure(figsize=(30, 6))
sns.barplot(x=city_distribution.index, y=city_distribution.values)
plt.xticks(rotation=90)
plt.title('Distribution of Restaurants by City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
#2(b) Distribution of restaurants by country
country_distribution = df['Country Code'].value_counts()
plt.figure(figsize=(18, 6))
sns.barplot(x=country_distribution.index, y=country_distribution.values)
plt.xticks(rotation=90)
plt.title('Distribution of Restaurants by Country')
plt.xlabel('Country Code')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
from scipy.stats import pearsonr

# Calculate correlation between latitude, longitude, and rating
latitude_rating_corr = pearsonr(df['Latitude'], df['Aggregate rating'])
longitude_rating_corr = pearsonr(df['Longitude'], df['Aggregate rating'])

print(f"Correlation between latitude and rating: {latitude_rating_corr[0]:.2f}")
print(f"Correlation between longitude and rating: {longitude_rating_corr[0]:.2f}")
```

## TO CONCLUDE :

~ Here is no linear relationship between the latitude of the restaurants and their ratings.
  A correlation of 0 indicates that changes in latitude do not predict changes in ratings at all.
  
~ There is a very weak negative linear relationship between the longitude of the restaurants and their ratings.
  A correlation of -0.12 suggests that as longitude increases or decreases, the rating tends to decrease slighty,
  but the relationship is very weak.
